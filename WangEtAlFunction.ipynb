# Wang Et Al Function 
*(used in Precipitate Growth Code)*

Demonstatres how both the conserved and non conserved variable can be combined together to use as order paramters to describe a microstructure. 

```
import numpy as np
from mpl_toolkits.mplot3d import Axes3D 
from matplotlib import pyplot as plt, cm
%matplotlib inline 
N = 141 

c = np.linspace(-0.2, 1.2, N)
phi = np.linspace(-0.2, 1.2, N) 

C, Phi = np.meshgrid(c,phi)

A = 1.0
B = 1.0
P = 1.0

def w(phi) :
    return phi*phi*phi * (10 - 15 * phi + 6 * phi*phi) 

WangFn = w(Phi)

for i in range (0,N):
    for j in range(0,N):
        if (WangFn[i][j] < 0) :
            WangFn[i][j] = 0.0
        elif (WangFn[i][j] > 1) :
            WangFn[i][j] = 1.0
        
F = A * C * C * (1 - WangFn) + B * (1 - C) * (1 - C) * WangFn + P * Phi * Phi * (1 - Phi) * (1 - Phi)        

fig = plt.figure(figsize=(11,7), dpi=100)
ax = fig.gca(projection='3d')
surf = ax.plot_surface(C, Phi, F,cmap=cm.gnuplot) 
```
