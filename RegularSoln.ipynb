# Regular Solution 
The expression of free energy change for a regular solution is as follows :

$\Delta$G = $\Omega$x(1-x) + RT[ xln(x) + (1-x)ln(1-x) ] 

where 

>$\Delta$H = $\Omega$x(1-x)

>$\Delta$S = RT[ xln(x) + (1-x)ln(1-x) ]

As from computational aspects it's better to work with non - dimensionalized entities, we divide the $\Delta$G expression by RT.

$\Delta$G / RT = $\Omega$[x(1-x)] / RT + xln(x) + (1-x)ln(1-x)

We can write the expression simplified as

$\Delta$G = $\alpha$x(1-x) + xln(x) + (1-x)ln(1-x)

where $\alpha$ can be termed as non dimensional regular solution parameter.

```
# importing libraries
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 

# defining composition as array x
x = np.zeros(1000)
x[0] = 0.001
x_step = 0.001

for i in range(0, 999, 1) : # storing the values
    x[i+1] = x[i] + x_step
```

Now putting the value of $\alpha$ = 0.1. As $\alpha$ is inversely proportional to T i.e temperature, small value of $\alpha$ means taking temperature very high. 

```
DH = np.zeros(1000) # non dimensionalized enthalpy change
DS = np.zeros(1000) # non dimensionalized entropy change
DG = np.zeros(1000) # non dimensionalized free energy change
a =  0.1 # defining a as alpha 

for i in range(0, 999, 1):
  DH[i] = a * x[i] * (1 - x[i])
  DS[i] = x[i] * np.log (x[i]) + (1 - x[i]) * np.log(1 - x[i])
  DG[i] = DH[i] + DS[i]

plt.plot(x, DH, color = "red")
plt.plot(x, DS, color = "green")
plt.plot(x, DG, color = "blue")
plt.xlabel ("Composition % of B in A")
plt.ylabel ("$\Delta$G / $\Delta$H / $\Delta$S ")
plt.title ("G/H/S vs X")
```

As seen in the plot above, the red curve is H vs X, green for S vs X and blue is G vs X. 
Since we chose temperature to be very high, the dominant contributor to G is the entropy part and hence the blue curve is downwards parabola. 

Now increasing the value of $\alpha$ = 1.5 i.e we are lowering the temperature. 

```
DH = np.zeros(1000) # non dimensionalized enthalpy change
DS = np.zeros(1000) # non dimensionalized entropy change
DG = np.zeros(1000) # non dimensionalized free energy change
a =  1.5 # defining a as alpha 

for i in range(0, 999, 1):
  DH[i] = a * x[i] * (1 - x[i])
  DS[i] = x[i] * np.log (x[i]) + (1 - x[i]) * np.log(1 - x[i])
  DG[i] = DH[i] + DS[i]

plt.plot(x, DH, color = "red")
plt.plot(x, DS, color = "green")
plt.plot(x, DG, color = "blue")
plt.xlabel ("Composition % of B in A")
plt.ylabel ("$\Delta$G / $\Delta$H / $\Delta$S ")
plt.title ("G/H/S vs X")
```

Here, as can been seen, due to lower temperature involved the H part becomes signigicantly positive and contributes to G. 

Hence the blue curve goes upward.

Now setting $\alpha$ = 4.5 i.e we are lowering the temperature significantly.

```
DH = np.zeros(1000) # non dimensionalized enthalpy change
DS = np.zeros(1000) # non dimensionalized entropy change
DG = np.zeros(1000) # non dimensionalized free energy change
a =  4.5 # defining a as alpha 

for i in range(0, 999, 1):
  DH[i] = a * x[i] * (1 - x[i])
  DS[i] = x[i] * np.log (x[i]) + (1 - x[i]) * np.log(1 - x[i])
  DG[i] = DH[i] + DS[i]

plt.plot(x, DH, color = "red")
plt.plot(x, DS, color = "green")
plt.plot(x, DG, color = "blue")
plt.xlabel ("Composition % of B in A")
plt.ylabel ("$\Delta$G / $\Delta$H / $\Delta$S ")
plt.title ("G/H/S vs X")
```

Here the blue curve forms a maxima due to the significant contributuion by H part. 
However, at the two ends, there are two minima wells forming. These wells will always be there, except at absolute zero, where entropy contributution becomes zero and the two minima wells disappear. 

Below, the code slowly increases $\alpha$ to observe the shift in the blue curve, i.e G.

```
for a in range(1,5,1) :    
    DH = np.zeros(1000)
    DS = np.zeros(1000)
    DG = np.zeros(1000)

    for i in range(0, 999, 1):
        DH[i] = a * x[i] * (1 - x[i])
        DS[i] = x[i] * np.log (x[i]) + (1 - x[i]) * np.log(1 - x[i])
        DG[i] = DH[i] + DS[i]

    plt.plot(x, DH, color = "red")
    plt.plot(x, DS, color = "green")
    plt.plot(x, DG, color = "blue")
    plt.xlabel ("Composition % of B in A")
    plt.ylabel ("$\Delta$G / $\Delta$H / $\Delta$S ")
    plt.title ("G/H/S vs X")
```

In the above plot we can see the shift of blue curve upwards to form a maxima from global minima and how the two minina wells are always there due to entropy part. 

The following curves demonstrate what happens when $\alpha$ = 0

```
DH = np.zeros(1000) # non dimensionalized enthalpy change
DS = np.zeros(1000) # non dimensionalized entropy change
DG = np.zeros(1000) # non dimensionalized free energy change
a =  0 # defining a as alpha 

for i in range(0, 999, 1):
  DH[i] = a * x[i] * (1 - x[i])
  DS[i] = x[i] * np.log (x[i]) + (1 - x[i]) * np.log(1 - x[i])
  DG[i] = DH[i] + DS[i]
plt.figure(0)
plt.plot(x, DH, color = "red")
plt.xlabel ("Composition % of B in A")
plt.ylabel ("$\Delta$H")
plt.title("$\Delta$H vs X")
plt.figure(1)
plt.plot(x, DS, color = "green")
plt.xlabel ("Composition % of B in A")
plt.ylabel ("$\Delta$S")
plt.title("$\Delta$S vs X")
plt.figure(2)
plt.plot(x, DG, color = "blue")
plt.xlabel ("Composition % of B in A")
plt.ylabel ("$\Delta$G")
plt.title("$\Delta$G vs X")
```

```
DH = np.zeros(1000) # non dimensionalized enthalpy change
DS = np.zeros(1000) # non dimensionalized entropy change
DG = np.zeros(1000) # non dimensionalized free energy change
a =  0.1 # defining a as alpha 
DG_Zero = np.zeros(1000)

for i in range(0, 999, 1):
  DH[i] = a * x[i] * (1 - x[i])
  DS[i] = x[i] * np.log (x[i]) + (1 - x[i]) * np.log(1 - x[i])
  DG[i] = DH[i] * 0.1 + DS[i] * 8.314 * 0

plt.plot(x, DG, color = "blue")
plt.xlabel ("Composition % of B in A")
plt.ylabel ("$\Delta$G / $\Delta$H / $\Delta$S ")
plt.title ("G vs X") 
```

As we can see here, at absolute zero, the two minima wells disappear completely as the entropy contribution becomes 0.
