#**Ideal Solution**

The following code plots G vs X i.e Free Energy vs Composition plot for the ideal solution in python.

We know the free energy change associated with ideal solution is 

$\Delta$G = RT [ xln(x) + (1 - x)ln(1 - x) ]  



```
# importing libraries
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Now we define the composition as an array named as `x` of required size and initialize it along with an array of $\Delta$G to store the final values. 

```
x = np.zeros(1000) # array of size 1000
x[0] = 0.001
step = 0.001 

for i in range(0, 999, 1) :
  x[i+1] = x[i] + step

deltaG = np.zeros(1000)  
```

In computational aspects, it's always better to plot and code the entities as their non dimensionalized versions. 
To make $\Delta$G non dimensional in the ideal solution expression, we divide it with "RT" to give the following expression :

$\Delta$G / RT = [ xln(x) + (1 - x)ln(1 - x) ] 

```
for i in range(0 , 999, 1) :
  deltaG[i] = x[i] * np.log (x[i]) + (1 - x[i]) * np.log(1 - x[i]) 
  # non dimensionalized version of free energy change
```

```
# Plotting the results 
plt.plot(x, deltaG)
plt.xlabel("Composition % of B in A")
plt.ylabel("Non Dimensionalized $\Delta$G")
plt.title("G vs X for Ideal Solution")
```

Now for the dimensionalized $\Delta$G, we need to multiply with universal gas constant R and the temperature we need the plot at. Code is shown below.

```
DG_300 = np.zeros(1000)  # array for dimensionalized $\Delta$G
    
for i in range(0,999,1):
    DG_300[i] = deltaG[i] * 8.314 * 300
# Plotting the results     
plt.plot(x, DG_300, color = 'red')
plt.xlabel("Composition % of B in A")
plt.ylabel("Dimensionalized $\Delta$G")
plt.title("G vs X for Ideal Solution")
```
